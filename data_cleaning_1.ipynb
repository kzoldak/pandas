__Write functions that check for certain characters or a word.__

````python
def check_for_characters(entries, characters=None):
    ''' 
    check_for_characters(entries, characters=None)
    
    PARAMETERS:
    -----------
    entries: list, list of strings to check if characters exists within any
                of the strings in the list of entries. 
    characters: None or list. None is default which uses all default characters
                listed below. You can pass a list of characters to limit which 
                ones you want the function to search for. 
    
                When  characters = None,
                Default characters are: 
                            ['!', '"', '#', '$', '%', '&', "'", 
                             '(', ')', '*', '+', ',', '-', '.', 
                             '/', ':', ';', '<', '=', '>', '?', 
                             '@', '[', '\\', ']', '^', '_', '`', 
                             '{', '|', '}', '~']
                             
                characters does not take words. 
    
    RETURNS:
    -----------
    Returns a list of results that contain one of the characters.
    
    NOTES:
    ----------
    This function's operation is to return entries with any of the 
    characters passed to it. So it has an OR operation functionality.
    Return any entry that has a '+' or a '-' character in it when
    characters=['+', '-']. If you only want cases where BOTH are 
    present in the entry, this is not the right function!
    
    EXAMPLE:
    ---------
    check_for_characters(entries=names, characters=None)
    check_for_characters(entries=names, characters=['+', '-'])
    check_for_characters(entries=names, characters='+')
    check_for_characters(entries=names, characters='\+')
    
    '''
    if characters is None:
        # use default chars
        import string
        chars = string.printable.split('Z')[1].split(' \t')[0] # string
        chars = list(chars)  # list of all possible characters
    else:
        chars = characters
    
    hits = []
    for entry in entries:
        if any(i in entry for i in tuple(chars)):
            hits.append(entry)
    return hits



def check_for_word(entries, word):   
    '''
    need to consider case sensitivity
    use case = True as default
    '''
    hits = []
    for entry in entries:
        if word in entry:
            hits.append(entry)
    return hits
    
````

````python
check_for_characters(entries=names, characters='+') # this works
````

````python
check_for_characters(entries=names, characters='\+') # this also works
````

````python
check_for_characters(entries=names, characters=['+', '-']) # this also works
````

````python
check_for_characters(entries=names, characters=' and ') # this does not work
````

__Build a funciton to replace '+' with 'and'__

````python
def replace(entries, val1, val2): 
    all_entries = []
    for entry in entries:
        if val1 in entry:
            all_entries.append(entry.replace(val1, val2))
    return all_entries

replace(names, ' + ', ' and ')
````

````python
replace = lambda entries,val1,val2: [entry.replace(val1,val2) for entry in entries if val1 in entry]

replace(names, ' + ', ' and ')
````

__The above two replace functions are equivalent.__

````python
import numpy as np
````

````python
# THIS ONE RETURNS ALL ENTRIES, INCLUDING THOSE THAT WERE NOT FIXED.
def replace(entries, val1, val2):
    out = [np.where(val1 in entry, entry.replace(val1, val2), entry).tolist() for entry in entries]
    return out

replace(names, ' + ', ' and ')
````
